### Método IQR

### Creamos un DataFrame temporal llamado 'temp'. Luego, iteramos sobre cada columna en 'data_clean' para identificar y etiquetar los valores atípicos (outliers) utilizando el método del rango intercuartílico (IQR). 
### Para cada columna, calculamos el primer cuartil (Q1) y el tercer cuartil (Q3). A partir de estos valores, determinamos el IQR y, con él, establecemos los umbrales inferior y superior para identificar outliers.
### Añadimos una nueva columna al DataFrame 'data_clean' con el sufijo '_outliers' que indica con valores booleanos (True/False) si un valor en la columna original es un outlier.
### Finalmente, imprimimos los valores calculados de Q1, Q3, IQR, y los umbrales para cada columna.


```python
temp = pd.DataFrame()

for col in data_clean.columns:
    column = data_clean[col]
 
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1
    
    umbral_inferior = Q1 - 1.5 * IQR
    umbral_superior = Q3 + 1.5 * IQR

    data_clean[col + '_outliers'] = (column < umbral_inferior) | (column > umbral_superior)

    #temp[col + '_outliers'] = ((column < umbral_inferior) | (column > umbral_superior))

    print(f'{Q1 =} {Q3=} {IQR =} {umbral_inferior = } {umbral_superior =}')
```

### Importamos el módulo stats de la biblioteca scipy.
### Inicializamos listas vacías 'lambdas' y 'nombre' para almacenar los valores de lambda y los nombres de las columnas, respectivamente. También creamos un nuevo DataFrame vacío llamado 'df_new'.
### Iteramos sobre cada columna en 'data_noout'. Para cada columna, aplicamos la transformación Box-Cox utilizando stats.boxcox() para normalizar la distribución de los datos. Almacenamos el valor de lambda y el nombre de la columna en las listas correspondientes. Luego, añadimos la columna transformada al DataFrame 'df_new'.
### Al final de este proceso, 'df_new' contendrá las versiones transformadas de las columnas de 'data_noout', y las listas 'lambdas' y 'nombre' tendrán los valores de lambda y los nombres de las columnas, respectivamente.


```python
from scipy import stats

lambdas = []
nombre = []
df_new = pd.DataFrame()

for x in data_noout.columns:
    df, lamb = stats.boxcox(data_noout[x])
    lambdas.append(lamb)
    nombre.append(x)
    df_new[x] = df
```

### Creamos un diccionario 'df_lambdas' con dos claves: 'nombre' y 'lambda'. Las claves se asocian con las listas 'nombre' y 'lambdas' que contienen los nombres de las columnas y los valores de lambda obtenidos de la transformación Box-Cox, respectivamente.
### Convertimos el diccionario 'df_lambdas' en un DataFrame para una representación tabular y fácil visualización de los nombres de las columnas y sus correspondientes valores de lambda.


```python
df_lambdas = {
    'nombre' : nombre,
    'lambda' : lambdas
}

df_lambdas = pd.DataFrame(df_lambdas)
df_lambdas
```

## Metodo del codo
### Importamos la clase KMeans de la biblioteca sklearn.cluster. KMeans es un algoritmo de clustering que nos permite agrupar datos en base a sus características. 

```python
from sklearn.cluster import KMeans
```

### Inicializamos dos listas vacías, 'inercia' y 'ks', para almacenar los valores de inercia y el número de clusters, respectivamente.
### Iteramos desde 1 hasta 60 para diferentes valores de 'k' (número de clusters). Para cada valor de 'k', creamos y ajustamos un modelo KMeans al DataFrame 'data_bc'.
### Después de ajustar el modelo, almacenamos el valor de inercia (que representa la suma de las distancias al cuadrado de cada punto a su centroide más cercano) en la lista 'inercia'. También almacenamos el valor actual de 'k' en la lista 'ks'.
### Esta iteración nos ayudará a determinar el número óptimo de clusters para nuestro conjunto de datos utilizando el método del codo, observando cómo cambia la inercia a medida que aumenta el número de clusters.


```python
inercia = []
ks = []

for k in range(1,60):
    modelo_iterado = KMeans(n_clusters = k, random_state = 5)
    modelo_iterado.fit(data_bc)
    inercia.append(modelo_iterado.inertia_)
    ks.append(k)
```

### Creamos un diccionario 'codo' con dos claves: 'inercia' y 'k'. Las claves se asocian con las listas 'inercia' y 'ks' que contienen los valores de inercia y el número de clusters obtenidos de la iteración anterior, respectivamente.
### Convertimos el diccionario 'codo' en un DataFrame para una representación tabular y fácil visualización de la relación entre el número de clusters y su correspondiente inercia.
### Este DataFrame nos ayudará a visualizar y determinar el punto de inflexión o "codo" donde la inercia comienza a estabilizarse, indicando el número óptimo de clusters para el conjunto de datos.


```python
codo = {
    'inercia': inercia,
    'k': ks
}

codo = pd.DataFrame(codo)
```

### Utilizamos la función scatterplot de seaborn para crear un gráfico de dispersión que muestra la relación entre el número de clusters 'k' y su correspondiente inercia. Cada punto en el gráfico representa un valor de 'k' y su inercia asociada.
### Luego, utilizamos la función lineplot de seaborn para trazar una línea que conecta estos puntos, facilitando la identificación del punto de inflexión o "codo" donde la inercia comienza a estabilizarse.
### Este gráfico nos ayudará a visualizar y determinar el número óptimo de clusters.


```python
sns.scatterplot(data=codo,x='k',y='inercia',markers=True)
sns.lineplot(data=codo, x='k', y='inercia',)
```

### Creamos un modelo KMeans con 6 clusters, utilizando un estado aleatorio de 10 para garantizar la reproducibilidad de los resultados. 
### Luego, ajustamos este modelo al DataFrame 'data_bc' que contiene nuestros datos transformados con Box-Cox. 
### Al final de este proceso, el modelo habrá segmentado los datos en 6 grupos homogéneos basados en las características de 'data_bc'.


```python
modelo = KMeans(n_clusters=6,random_state = 10)
modelo.fit(data_bc)
```

### Añadimos una nueva columna llamada cluster a data_noout. Esta columna almacenará las etiquetas de cluster asignadas a cada registro por el modelo KMeans que ajustamos anteriormente.
### De esta manera, cada registro en data_noout ahora tiene una etiqueta de cluster que indica a qué grupo pertenece.


```python
data_noout['cluster'] = modelo.labels_
data_noout
```
